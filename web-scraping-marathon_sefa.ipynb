```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
full_2016 = pd.read_csv('data/full_2016.csv')
full_2017 = pd.read_csv('data/full_2017.csv')
full_2018 = pd.read_csv('data/full_2018.csv')
full_2019 = pd.read_csv('data/full_2019.csv')
half_2016 = pd.read_csv('data/half_2016.csv')
half_2017 = pd.read_csv('data/half_2017.csv')
half_2018 = pd.read_csv('data/half_2018.csv')
half_2019 = pd.read_csv('data/half_2019.csv')
```

```python
marathon_full  = pd.merge(full_2016,full_2017, on=full_2016.columns.tolist(), how='outer')
marathon_full = pd.merge(marathon_full,full_2018, on=marathon_full.columns.tolist(), how='outer')
marathon_full = pd.merge(marathon_full,full_2019, on=marathon_full.columns.tolist(), how='outer')

marathon_half = pd.merge(half_2016,half_2017, on=full_2016.columns.tolist(), how='outer')
marathon_half = pd.merge(marathon_half,half_2018, on=marathon_half.columns.tolist(), how='outer')
marathon_half = pd.merge(marathon_half,half_2019, on=marathon_half.columns.tolist(), how='outer')
```

```python
mar_full_year = marathon_full[['Bib', 'year']].groupby('year').count()
mar_full_year = mar_full_year.rename(columns={'Bib': 'all_runner'})
mar_full_year['non_comp_runner'] = marathon_full[marathon_full.Overall==99999][['Bib', 'year']].groupby('year').count().Bib
mar_full_year['comp_runner'] = marathon_full[marathon_full.Overall!=99999][['Bib', 'year']].groupby('year').count().Bib
mar_full_year['perc_completed'] = 100 - 100*(mar_full_year.non_comp_runner/mar_full_year.all_runner)
mar_full_year['category'] = 'Marathon'
mar_full_year = mar_full_year.reset_index()
mar_full_year
```

It is interesting that in 2018, the number of runners who did not complete the marathon is so low. What about the number of all runners?

```python
mar_half_year = marathon_half[['Bib', 'year']].groupby('year').count()
mar_half_year = mar_half_year.rename(columns={'Bib': 'all_runner'})
mar_half_year['non_comp_runner'] = marathon_half[marathon_half.Overall==99999][['Bib', 'year']].groupby('year').count().Bib
mar_half_year['comp_runner'] = marathon_half[marathon_half.Overall!=99999][['Bib', 'year']].groupby('year').count().Bib
mar_half_year.non_comp_runner = mar_half_year.non_comp_runner.fillna(int(0)).astype(int)
mar_half_year['perc_completed'] = 100 - 100*(mar_half_year.non_comp_runner/mar_half_year.all_runner)
mar_half_year['category'] = 'Half-Marathon'
mar_half_year = mar_half_year.reset_index()
mar_half_year
```

### For half marathons:

```python
duplicated_half = marathon_half[marathon_half.duplicated(subset=['Name', 'year'],keep=False)].sort_values(by=['year','Name']).Name.value_counts().index.tolist()
len(duplicated_half)
```

We have 565 runners who compete with at least one runner having the exact same name! (duplicates by subsetting name and year at the same time).

```python
frequent_half = marathon_half[marathon_half.duplicated(subset=['Name'],keep=False)].Name.value_counts().index.tolist()
len(frequent_half)
```

We have 8201 runners whose name is seen at least two times! (duplicates by subsetting name). 8201-565 is the number of runners who ran at leas two years and has unique names.

### For full marathons:

```python
duplicated_full = marathon_full[marathon_full.duplicated(subset=['Name', 'year'],keep=False)].sort_values(by=['year','Name']).Name.value_counts().index.tolist()
len(duplicated_full)
```

We have 23 runners who compete with at least one runner having the exact same name! (duplicates by subsetting name and year at the same time).

```python
frequent_full = marathon_full[marathon_full.duplicated(subset=['Name'],keep=False)].Name.value_counts().index.tolist()
len(frequent_full)
```

We have 574 runners whose name is seen at least two times! (duplicates by subsetting name). 574-23 is the number of runners who ran at leas two years and has unique names.

```python
frequent_full_result = marathon_full[marathon_full.Name.isin(frequent_full) & ~marathon_full.Name.isin(duplicated_full)].sort_values(by=['Name', 'year']).reset_index()
frequent_half_result = marathon_half[marathon_half.Name.isin(frequent_half) & ~marathon_half.Name.isin(duplicated_half)].sort_values(by=['Name', 'year']).reset_index()
frequent_full_result.Time = pd.to_timedelta(frequent_full_result.Time)
frequent_half_result.Time = pd.to_timedelta(frequent_half_result.Time)

frequent_full_result['idle_time'] = np.where(frequent_full_result.Name==frequent_full_result.Name.shift(),
                                            frequent_full_result.year-frequent_full_result.year.shift(),
                                            None)
frequent_full_result['performance_change'] = np.where(frequent_full_result.idle_time.isna(),
                                                      None,
                                                      (frequent_full_result.Time.dt.total_seconds()-frequent_full_result.Time.dt.total_seconds().shift())/60)

frequent_half_result['idle_time'] = np.where(frequent_half_result.Name==frequent_half_result.Name.shift(),
                                            frequent_half_result.year-frequent_half_result.year.shift(),
                                            None)
frequent_half_result['performance_change'] = np.where(frequent_half_result.idle_time.isna(),
                                                      None,
                                                      (frequent_half_result.Time.dt.total_seconds()-frequent_half_result.Time.dt.total_seconds().shift())/60)
```

```python
plt.scatter(frequent_full_result.idle_time,frequent_full_result.performance_change)
plt.ylabel('Performance change in minutes', fontsize=13)
plt.xlabel('Year elapsed from the last race', fontsize=15)
plt.xlim([0,4])
plt.title('Full Marathon', fontsize=20)
```

```python
plt.scatter(frequent_half_result.idle_time,frequent_half_result.performance_change, color='red')
plt.ylabel('Performance change in minutes', fontsize=13)
plt.xlabel('Year elapsed from the last race', fontsize=15)
plt.xlim([0,4])
plt.title('Half Marathon', fontsize=20)
```

```python
frequent_full_result_perf = frequent_full_result[['idle_time', 'performance_change']].dropna()
frequent_half_result_perf = frequent_half_result[['idle_time', 'performance_change']].dropna()
frequent_full_result_perf.performance_change = frequent_full_result_perf.performance_change.astype(float)
frequent_half_result_perf.performance_change = frequent_half_result_perf.performance_change.astype(float)

frequent_full_result_perf.groupby('idle_time').describe()
```
